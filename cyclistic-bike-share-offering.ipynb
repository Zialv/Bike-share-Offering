```python
def create_dict(df, column1, column2):
    # Decide to choose the most frequent column2' values to be the main values
    # Count the frequency of column2' values group by column 1.
    # Then create a dictionary include all most frequent column2' value for column 1 
    dict_1 = df.groupby(column1)[column2].apply(lambda x:x.value_counts().idxmax()).to_dict()
    # Create a new column in data frame, map this data frame to above dictionary by column 1
    df['new_column'] = df[column1].map(dict_1)
    # Make column 2 = new_column
    df[column2] = df['new_column']
    # Then drop new_column 
    df = df.drop(columns=['new_column'])
    # Create dictionary 
    df.drop_duplicates(inplace=True)
    df.set_index(column1,inplace=True)
    dict_2 = df.to_dict()
    dict_3 = dict_2[column2]
    return dict_3
```

<div style="background-color:#F8F5D5; padding: 20px;">
    <h2 style="font-size:20px"><b>Create a function to cacualate ride_length</b></h2>
        <ul style="font-size:14px">
        <li>This will be the length of the ride in minutes, hours and be cacualted by subtracting end time from start time.</li>
        <li>After caculation, we decide to drop all record have time less than 1 minutes (As 1 minutes is too short for a ride)</li>
        </ul>
</div>

```python
def create_ride_length_feat(df):
    # Create ride_length in hours:
    df['no_of_hours'] = (df['ended_at'] - df['started_at']).abs().dt.total_seconds() / 3600
    #Create ride_length in minutes: 
    df['no_of_minutes'] = (df['ended_at'] - df['started_at']).abs().dt.total_seconds() / 60
    df = df[df['no_of_minutes'] > 1]
    return df
```

<div style="background-color:#F8F5D5; padding: 20px;">
    <h2 style="font-size:20px"><b>Create a function to retrieve hour, date, month, year from start time column</b></h2>
</div>

```python
def create_time_feat(df): 
    # Create Hour column
    df['hour'] = df['started_at'].dt.hour
    # Create day of week column
    df['day_of_week'] = df['started_at'].dt.dayofweek + 1 #Monday = 0 in default 
    # Create month column 
    df['month'] = df['started_at'].dt.month 
    # Create year column 
    df['year'] = df['started_at'].dt.year
    return df
```

<div style="background-color:#F8F5D5; padding: 20px;">
    <h2 style="font-size:20px"><b>Create a function to cacualate distance</b></h2>
        <ul style="font-size:14px">
            <li>The Haversine formula calculates the distance between two points on the Earth's surface given their coordinates. It takes into account the Earth's curvature to provide an accurate distance.</li>
            <li>This feature is only relative because the longitude and latitude of the starting and ending points do not represent the exact distance the customer has cycled.</li>
            <li>We create a function to cacualte the distance of start and end points based on longtitude and latitude. </li>
            <li>However, there are cases where customers start and end at the same station, resulting in a calculated distance of zero. In such instances, we will replace zero distances with the average distance traveled by casual or member customers, depending on the customer type.</li>
        </ul>
</div>

```python
def calculate_distance(row):
    # Radius of the earth in km
    R = 6371.0

    lat1 = np.radians(row['start_lat'])
    lon1 = np.radians(row['start_lng'])
    lat2 = np.radians(row['end_lat'])
    lon2 = np.radians(row['end_lng'])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    # Convert to kilometers
    distance = R * c

    return distance
# # Calculate the distance
# trip_data['distance_km'] = trip_data.apply(calculate_distance, axis=1)
```

```python
def create_distance_feat(df):
        # Calculate the distance
    df['distance_km'] = calculate_distance(trip_data)
    filtered_df = df[df['distance_km'] != 0]
    mean_casual = filtered_df.groupby('member_casual')['distance_km'].mean().get(key='casual')
    mean_member = filtered_df.groupby('member_casual')['distance_km'].mean().get(key='member')
    df['distance_km'] = np.where((df['member_casual'] == 'member') & (df['distance_km'] == 0), mean_member, df['distance_km'])
    df['distance_km'] = np.where((df['member_casual'] == 'casual') & (df['distance_km'] == 0), mean_casual, df['distance_km'])
    return df
```

<div style="background-color:#F8F5D5; padding: 20px;">
    <h2 style="font-size:20px"><b>Create length_lables, ride_length_category and distance_category</b></h2>
</div>

```python
def create_category_feat(df):
    # Create length labels
    df['duration_label'] = np.where(df['no_of_hours'] <= 24, 'in day', 'long days')
    # Create Duration Category in Hours
    length_bins = [0, 5, 15, 30, 60, np.inf]
    length_names = ['<5', '<15', '<30', '<60', '60+']
    df['duration_category'] =pd.cut(df['no_of_minutes'], length_bins, labels = length_names)
    # Create distance category 
    bins = [0, 1, 5, 10, np.inf ]
    names = ['<1', '<5', '<10', '10+']
    df['distance category'] = pd.cut(df['distance_km'], bins, labels=names)
    return df
```
